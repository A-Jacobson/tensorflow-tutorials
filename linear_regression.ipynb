## Linear Regression

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

#### Create 100 phony x, y data points in NumPy in the form y = Wx+b. 
#### concretely: y = 0.1x + 0.3

```python
x_data = np.random.rand(100).astype(np.float32)
y_data = 0.1 * x_data + 0.3
```

```python
print x_data.shape, y_data.shape
```

```python
plt.scatter(x_data, y_data)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0,1])
plt.xlim([0,1])
plt.title('y = 0.1x + b');
```

#### Try to find values for W and b that compute y_data = W * x + b
Given X_data, try to find a value to multiply by (W) and a value to add to x_data to get b

```python
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b # (1 x n) * (n * 1) + 1 = 1 <<--- dimensions

loss = tf.reduce_mean(tf.square(y - y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)
```

#### before starting, init the viariables. We will 'run' this first.

```python
init = tf.initialize_all_variables()
```

```python
x_data.dtype
```

#### Fit the line.

```python
with tf.Session() as sess:
    sess.run(init) # initialize variables
    for step in range(200): # run graph 200 times
        sess.run([train])
        if step % 20 == 0:
            print "step: %d, W: %f, b: %f, loss: %f" % (step, sess.run(W), sess.run(b), sess.run(loss))
            y_pred = sess.run(y)
```

```python
plt.scatter(x_data, y_data)
plt.plot(x_data, y_pred)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0,1])
plt.xlim([0,1])
plt.title('y = 0.1x + b');
```

### nonlinear function

```python
x_data = np.random.rand(100).astype(np.float32)
y_data = x_data**3  + 0.3
```

```python
plt.scatter(x_data, y_data)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0,1])
plt.xlim([0,1])
plt.title('y = x_data**3  + 0.3');
```

```python
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b # (1 x n) * (n * 1) + 1 = 1 <<--- dimensions

loss = tf.reduce_mean(tf.square(y - y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)

init = tf.initialize_all_variables()


with tf.Session() as sess:
    sess.run(init) # initialize variables
    for step in range(200): # run graph 200 times
        sess.run([train])
        if step % 20 == 0:
            print "step: %d, W: %f, b: %f, loss: %f" % (step, sess.run(W), sess.run(b), sess.run(loss))
            y_pred = sess.run(y)
```

```python
plt.scatter(x_data, y_data)
plt.plot(x_data, y_pred)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0,1])
plt.xlim([0,1])
plt.title('y = x_data**3  + 0.3');
```
